```python
import matplotlib.pyplot as plt
import json
import numpy as np

# I don't know why we need this but we need it otherwise, font.size won't work.
fig = plt.figure(figsize =(12, 6))
fig.show()

plt.rcParams.update({'font.size': 16})

with open("result_gcloud.json", "r") as f:
    d = json.load(f)
```

```python
testsuite = []
for key in d.keys():
    if key[-3:] == "_sc":
        continue
    testsuite.append(key)
print(testsuite)
```

```python
labels = testsuite
byte = [np.mean(d[key]["time"]) for key in testsuite]
source = [np.mean(d[key+"_sc"]["time"]) for key in testsuite]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig = plt.figure(figsize =(12, 6))
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
rects1 = ax.bar(x - width/2, byte, width, label='bytecode')
rects2 = ax.bar(x + width/2, source, width, label='sourcecode')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Speed (sec)')
ax.set_title('Speed of deserialization')
ax.set_xticks(x) # values
ax.set_xticklabels(labels, rotation=(45)) # labels
ax.legend()
# plt.yticks(np.arange(0, 0.12, 0.02))

fig.tight_layout()

plt.show()
```
