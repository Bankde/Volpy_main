```python
import matplotlib.pyplot as plt
import json
import numpy as np

with open("result.json", "r") as f:
    d = json.load(f)
```

```python
testsuite = []
for key in d.keys():
    if key[-3:] == "_sc":
        continue
    testsuite.append(key)
print(testsuite)
```

```python
labels = testsuite
byte = [d[key]["time"] for key in testsuite]
source = [d[key+"_sc"]["time"] for key in testsuite]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, byte, width, label='bytecode')
rects2 = ax.bar(x + width/2, source, width, label='sourcecode')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Speed (sec)')
ax.set_title('Speed of deserialization')
ax.set_xticks(x) # values
ax.set_xticklabels(labels) # labels
ax.legend()

fig.tight_layout()

plt.show()
```
