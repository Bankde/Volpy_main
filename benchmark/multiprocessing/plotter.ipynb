```python
import os, sys
import re
import datetime
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

plt.rcParams.update({'font.size': 14})
```

```python
import json
f = open('result.json')
raw_data = json.load(f)
```

```python
data = []
data.append(raw_data["pythonNormal"])
data.append(raw_data["pythonThread_4"])
data.append(raw_data["pythonMultiProc_grpc"])
data.append(raw_data["js_worker_1"])
js_worker_4_time = []
for t in raw_data["js_worker_4"]:
    js_worker_4_time.append(max(t))
data.append(js_worker_4_time)

fig = plt.figure(figsize =(12, 6))
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(data)
ax.set_title('Time taken to run monte carlo pi (1M loop)')
xtick = ["Python regular", "Python 4-thread-pool", "Python 4-process-pool", "Pyodide 1-worker", "Pyodide 4-worker"]
ax.set_xticks(range(1,1+len(xtick)), xtick, rotation=20)
ax.set_ylabel('time (ms)')
plt.axvline(x=3.5)
fig.show()
```
